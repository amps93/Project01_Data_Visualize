## 정규표현식

```python
df['Country'].unique()
```

```python
df2 = df[['Country', 'Yearly Compensation']]
```

```python
df2['Yearly Compensation'].describe()
```

```python
#연봉 데이터 salary에 저장
salary = list(df2['Yearly Compensation'].unique())
print(salary)
```

```python
# 정규표현식 활용해 특수문자 제거
import re

numbers = re.sub(',','', salary[0])
numbers = re.findall('\d+', numbers)
list(map(int,numbers))
```

```python
manu_sal=[]
for i in range(len(salary)):
    numbers = re.sub(',','', salary[i])
    numbers = re.findall('\d+', numbers)
    manu_sal.append(list(map(int,numbers)))
# pd.DataFrame(manu_sal)
```

```python
#범위 숫자 두개를 합산 후 2로 나누어 한개의 값으로 만들고 salary에 저장
salary2 = []
for i in range(0,12):
    for j in range(0,1):
        a = int((manu_sal[i][j]+manu_sal[i][1])/2)+1
        salary2.append(a)
salary2.append(500000)
for i in range(13,25):
    for j in range(0,1):
        a = int((manu_sal[i][j]+manu_sal[i][1])/2)+1
        salary2.append(a)
print(salary2)
```

```python
#데이터 프레임에 연봉 데이터 변경
for i in range(0,25):
    df['Yearly Compensation'] = df['Yearly Compensation'].replace(salary[i],salary2[i])
```

```python
def get_score(v):
    if v == 'Korea':
        score = 1.2
    elif v == 'US':
        score = 1
    elif v == 'India':
        score = 0.6
    else:
        score = 0
    return score

df_ex["New_col"] = df_ex["Country"].apply(lambda v: get_score(v))
df_ex
```
